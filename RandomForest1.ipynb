```python
import pandas as pd
import numpy as np

with open('C:/Users/Aakarsh/Downloads/SURGE/FileNo_1_Stable_Qa475slpm.txt', 'r') as file:
   data=file.readlines()
   
Time1 = []
Pressure1 = []

for line in data:
    values = line.strip().split(',')
    Time1.append(values[0])
    Pressure1.append(values[6])

start_index = 0
end_index = 3276800
step = 500
mean_values1 = []
rms_values1 = []
sd_values1 = []

for i in range(start_index, end_index, step):
    mean = np.mean(np.array(Pressure1[i:i+step]).astype(float))
    rms= np.sqrt(np.mean(np.square(np.array(Pressure1[i:i+step]).astype(float))))
    sd= np.std(np.array(Pressure1[i:i+step]).astype(float)) 
    mean_values1.append(mean)
    rms_values1.append(rms)
    sd_values1.append(sd)

df1 = pd.DataFrame({'Mean': mean_values1,'RMS': rms_values1,'SD': sd_values1, 'Result': 0})

with open('C:/Users/Aakarsh/Downloads/SURGE/FileNo_6_Unstable_Qa600slpm.txt', 'r') as file:
   data=file.readlines()
   
Time2 = []
Pressure2 = []

for line in data:
    values = line.strip().split(',')
    Time2.append(values[0])
    Pressure2.append(values[6])

mean_values2 = []
rms_values2 = []
sd_values2 = []    

for i in range(start_index, end_index, step):
    mean = np.mean(np.array(Pressure2[i:i+step]).astype(float))
    rms= np.sqrt(np.mean(np.square(np.array(Pressure2[i:i+step]).astype(float))))
    sd= np.std(np.array(Pressure2[i:i+step]).astype(float)) 
    mean_values2.append(mean)
    rms_values2.append(rms)
    sd_values2.append(sd)

df2 = pd.DataFrame({'Mean': mean_values2,'RMS': rms_values2,'SD': sd_values2, 'Result': 1})

df = pd.concat([df1, df2], ignore_index=True)
shuffled_df = df.sample(frac=1, random_state=42)  # Shuffle the DataFrame using a random seed (random_state)
shuffled_df = shuffled_df.reset_index(drop=True)
df= shuffled_df
# df = df.apply(np.random.permutation)
x = df[['Mean','RMS','SD']].values
y = df['Result'].values
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score
accuracy = accuracy_score(y_test, y_pred)
precision= precision_score(y_test,y_pred)
recall = recall_score(y_test, y_pred)
F1score= f1_score(y_test,y_pred)
print(accuracy)
print(precision)
print(recall)
print(F1score)
```
